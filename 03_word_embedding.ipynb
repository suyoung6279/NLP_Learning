# WordEmbedding 기반 텍스트 분석 연습

1. 적절한 데이터셋을 찾거나 생성하고, 적절한 전처리를 진행한다. (01_preprocessing.ipynb)
2. Word Embedding model를 이용하여 벡터화한다.
3. Cosine Similarity를 계산하여 입력된 문자열의 긍/부정을 판단한다.

```python
import pandas as pd
df = pd.read_csv('실습_전처리.csv',encoding='utf-8')
df = df.dropna()
preprocessed_sentences = df.values.flatten()
preprocessed_sentences = [str(sentence).split() for sentence in preprocessed_sentences]
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=preprocessed_sentences,
    vector_size=100,
    window=5,
    min_count=5,
    sg=0
)
```

```python
model.wv.vectors.shape
```

```python
pd.DataFrame(model.wv.vectors, index=model.wv.index_to_key).head(30)
```

```python
model.wv.most_similar('회계사')
```

```python
model.wv.similarity('회계사', '직업')
```

```python
model.wv.save_word2vec_format('comment_kr_w2v')
```

```python
from gensim.models import KeyedVectors

load_model = KeyedVectors.load_word2vec_format('comment_kr_w2v')
```

```python
load_model.most_similar('직업')
```
