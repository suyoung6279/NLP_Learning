# BoW 기반 텍스트 분석 연습

1. 적절한 데이터셋을 찾거나 생성하고, 적절한 전처리를 진행한다. (01_preprocessing.ipynb)
2. TF-IDF Vectorizer를 이용하여 벡터화한다.
3. Cosine Similarity를 계산하여 입력된 문자열의 긍/부정을 판단한다.

```python
import pandas as pd
df = pd.read_csv('실습_전처리.csv',encoding='utf-8')
df = df.dropna()
preprocessed_sentences = df.values.flatten()
preprocessed_sentences
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(
    ngram_range=(1,2),
    max_df=1.0,
    min_df=2
)

comment_vecs = tfidf_vectorizer.fit_transform(preprocessed_sentences)
```

```python
tfidf_vectorizer.vocabulary_
```

```python
tfidf_vectorizer.get_feature_names_out()
```

```python
import pandas as pd
vocab = sorted(tfidf_vectorizer.vocabulary_.items(),key=lambda x : x[1])
vocab_df = pd.DataFrame(vocab,columns=['word','index'])
```

```python
vocab_df
```

```python
pd.DataFrame(comment_vecs.toarray(),columns=tfidf_vectorizer.get_feature_names_out())
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=2,
    max_iter=5000,
    random_state=0
)

news_label = kmeans.fit_predict(comment_vecs)
```

```python
center = kmeans.cluster_centers_
center_arg_idx = center.argsort()[:,::-1]
idx_top_20 = center_arg_idx[:,:40]
feature_names = tfidf_vectorizer.get_feature_names_out()
feature_names[idx_top_20]
```

```python
df = pd.DataFrame(preprocessed_sentences,columns=['댓글'])
df['label'] = kmeans.fit_predict(comment_vecs)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

base_idx = 214
baseline_vector = comment_vecs[base_idx]
baseline_vector

similarities = cosine_similarity(comment_vecs, baseline_vector).flatten()

df['base_similarity'] = similarities
print(df.iloc[base_idx]) 
```

```python
df.sort_values(by='base_similarity',ascending=False)
```
